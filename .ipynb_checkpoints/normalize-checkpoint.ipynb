```python
import torch
import torch.nn as nn
```

## batch norm

```python
bn2 = nn.BatchNorm2d(num_features=3)
print(bn2.weight)
print(bn2.bias)
print(bn2.weight.shape)
print(bn2.bias.shape)
```

## layer norm

```python
ln = nn.LayerNorm(normalized_shape=[3,128])
print(ln.weight)
print(ln.bias)
print(ln.weight.shape)
print(ln.bias.shape)
```
